```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)
```

```python
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
```

## K-최근접 이웃의 한계
회귀시 근처에 있는 평균을 구하기 때문에 범위를 벗어난 값은 오차가 있다.

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor(n_neighbors=3)

knr.fit(train_input, train_target)

print(knr.predict([[50]])) # 50cm 농어는 원래 1033 보다 더 높게 나와야 함
```

```python
import matplotlib.pyplot as plt

pre = 100
distance, indexes = knr.kneighbors([[pre]])

plt.scatter(train_input, train_target)
plt.scatter(train_input[indexes], train_target[indexes], marker='D')
plt.scatter(pre, knr.predict([[pre]]), marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

## 선형회귀

과소 적합이 일어남
(이유 : 훈련 세트의 점수가 높지않음)

아래 그래프를 볼떄 무게가 0g 이하 일 수도 있다는 문제가 있음

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(train_input, train_target)

print(lr.predict([[50]]))
```

```python
print(lr.coef_, lr.intercept_) # 각각 기울기와 상수항이며 모델 파라미터임
```

```python
plt.scatter(train_input, train_target)
plt.plot([15, 50], [15*lr.coef_ + lr.intercept_, 50*lr.coef_+lr.intercept_])
plt.scatter(50, lr.predict([[50]]), marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

## 다항회귀
2차 방정식 그래프를 만들려면 특성의 제곱한 항이 훈련 세트에 추가 되어야함

2차 방정식이라고 선형 회귀가 아님
(이유: x^2을 다른 특성으로 바꿀 수 있음)

과소 적합이 약간 남아있음


```python
train_poly = np.column_stack((train_input ** 2, train_input))
test_poly = np.column_stack((test_input ** 2, test_input))
```

```python
lr = LinearRegression()
lr.fit(train_poly, train_target)

print(lr.predict([[50**2, 50]]))
```

```python
print(lr.coef_, lr.intercept_) # 각각 기울기와 상수항이며 모델 파라미터임
```

```python
point = np.arange(15, 51)

plt.scatter(train_input, train_target)
plt.plot(point, lr.coef_[0]*point**2 + lr.coef_[1]*point + lr.intercept_)
plt.scatter(50, lr.predict([[50**2, 50]]), marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
```
